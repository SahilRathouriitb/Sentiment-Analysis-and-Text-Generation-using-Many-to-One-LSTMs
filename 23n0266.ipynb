TODO 3: Complete the function ```complete_cost``` function below. Refer to Problem statement for the cost (error) function for linear regression.

Here, ```X``` : np.array of x-coordinates of points

```y``` : np.array of y-coordinates of points

```theta``` : np.array of parameters

````python
def compute_cost(X, y, theta):
    m = len(y)
    X_b = np.c_[np.ones((m, 1)), X]  # Add bias term (x0 = 1)
    # since our best fit line is going to be of the form y=mx+b, we have the initial value of m and b , X_b has X values and a column with all 1's
    
    X_b[:,1]=theta[0]*X_b[:,1]
    X_b[:,0]=theta[1]*X_b[:,0]
    Y_pred=X_b[:,1]+X_b[:,0]
    Y_pred=Y_pred.reshape(100,1)
    z=(Y_pred-y)*(Y_pred-y)
    cost=(z.sum())*(1/(2*m))
    
    
    return cost
````

TODO 4: Implement Gradient Descent for linear regression according to algorithm given in the Problem Statement. Your only task is to update theta properly by finding the gradient correctly.

````python
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    X_b = np.c_[np.ones((m, 1)), X]  # Add bias term (x0 = 1)
    cost_history = []

    for i in range(0,iterations):
        #Complete TODO 4
        theta=theta-(learning_rate/m)*np.dot(np.transpose(X_b),np.dot(X_b,theta)-y)
        # np.dot is for matrix multiplication, the usual * wont work because it it multiplies elements side by side in similar shaped arrays 
        cost = compute_cost(X, y, theta)
        cost_history.append(cost)
        
       

    return  theta,cost_history
````
